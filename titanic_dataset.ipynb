```python
data["age"]= data["age"].fillna(data["age"].median())
data["embarked"]= data["embarked"].fillna(data["embarked"].mode()[0])
data["deck"]= data["deck"].fillna(data["deck"].mode()[0])
data["embark_town"]= data["embark_town"].fillna(data["embark_town"].mode()[0])
data.drop(["class","who","adult_male","alive"], axis=1, inplace=True, errors='ignore')
```

```python
data.columns
```

```python
data["age_bin"]= pd.cut(data["age"], bins=[0,12,18,35,60,80], labels= ["child", "teen", "young_adult", "elder", "senior_citizen"]) 
```

```python
data["family_size"]= data["sibsp"]+data["parch"]+1
data["is_alone"]= (data["family_size"]==1).astype(int)
```

```python
data["fare_bin"]= pd.qcut(data["fare"], q=4, labels= ["low", "medium", "high", "veryhigh"])
```

```python
data.info()
```

```python
categorical_cols= ["sex","embarked", "deck", "embark_town", "age_bin", "fare_bin"]
data= pd.get_dummies(data, columns= categorical_cols, drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
num_cols= ["age", "sibsp", "parch", "fare", "family_size"]
data[num_cols]= scaler.fit_transform(data[num_cols])
```

```python
data.head()
data.to_csv("titanic_cleaned.csv", index=False)
```
